```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the data
df = pd.read_csv('../data/synthetic_health_data.csv')

# 2. Descriptive Statistics
# This shows the "shape" of your customer base (e.g., average age, stress levels)
print("--- Data Summary ---")
print(df.describe())

# 3. Age Distribution
# Helps the IA understand their primary demographic
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='teal')
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 4. Correlation Heatmap (Crucial for KDM)
# This proves your "Business Logic" is working. 
# We expect to see a relationship between stress_level and interest_OSP.
plt.figure(figsize=(12, 8))
correlation_matrix = df.drop('user_id', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Health Metrics vs. Product Interest')
plt.show()

# 5. Visualizing the IA's Opportunity
# Scatter plot of Age vs. Ambrotose Interest
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='interest_Ambrotose', alpha=0.6)
plt.title('Age vs. Interest in Ambrotose 2x')
plt.show()
```
