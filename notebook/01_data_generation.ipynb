```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)
n_users = 1000

# 1. Generate Basic Profiles 
data = {
    'user_id': range(1, n_users + 1),
    'age': np.random.randint(20, 75, n_users),
    'stress_level': np.random.randint(1, 11, n_users),  # 1-10 scale
    'sleep_quality': np.random.randint(1, 11, n_users), # 1-10 scale
}

df = pd.DataFrame(data)

# 2. Add Business Logic (Creating "Personas") [cite: 28]
# Example: High stress usually correlates with lower sleep and interest in OSP
df['interest_OSP'] = df['stress_level'].apply(lambda x: np.random.randint(5, 11) if x > 7 else np.random.randint(1, 6))

# Example: Older age groups might show more interest in Ambrotose (Cellular health)
df['interest_Ambrotose'] = df['age'].apply(lambda x: np.random.randint(6, 11) if x > 50 else np.random.randint(1, 7))

# 3. Generate RFM Metrics 
df['recency_days'] = np.random.randint(1, 365, n_users)
df['frequency_visits'] = np.random.randint(1, 50, n_users)

# 4. Save to your 'data/' folder [cite: 129]
df.to_csv('../data/synthetic_health_data.csv', index=False)
print("Synthetic dataset created successfully!")
```
